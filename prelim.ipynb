```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
climate_dat = pd.read_csv("./data/climate-daily.csv")
climate_dat.drop(["x","y"], axis=1, inplace=True)

climate_dat.head()
```

```python
climate_dat["OTHER_PRECIPITATE"] = climate_dat["TOTAL_PRECIPITATION"] - climate_dat["TOTAL_RAIN"]
```

```python
sept_min_temps = climate_dat.loc[(climate_dat["LOCAL_MONTH"] == 1), ["LOCAL_YEAR", "OTHER_PRECIPITATE"]].groupby("LOCAL_YEAR").sum()
fig, ax = plt.subplots()
ax.plot(sept_min_temps.index[:-1], sept_min_temps[:-1])
ax.set_ylabel('Total September Precipitation')
ax.set_xlabel('Year')
# ax.set_ylim([-5,15])
plt.show()
```
